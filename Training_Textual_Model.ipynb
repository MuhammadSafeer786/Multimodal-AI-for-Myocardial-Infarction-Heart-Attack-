# Encoding Dataset

```python
df['gender'] = df['gender'].apply(lambda x: 0 if x == 'Male' else 1) # Male,0 and Female,1
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
df['medications'] = df['medications'].apply(lambda x: model.encode(x))
df['diagnosis'] = df['diagnosis'].apply(lambda x: model.encode(x))
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Combine embeddings
df['combined_embedding'] = df.apply(
    lambda row: np.concatenate([row['medications'], row['diagnosis']]), axis=1
)

additional_features = df[['gender', 'age', 'Troponin T(ng/ML)', 'emergency']].values
combined_features = np.hstack([np.vstack(df['combined_embedding']), additional_features])

target = df['abnormal'].values

X_train, X_test, y_train, y_test = train_test_split(combined_features, target, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

accuracy = rf_model.score(X_test, y_test)
print("Accuracy:", accuracy)
```
