## General Preprocessing

### Removing and Renaming some columns

```python
merged_df = merged_df.drop(['hadm_id', 'specimen_id', 'comments'], axis=1)
```

```python
merged_df = merged_df.rename(columns={'valuenum': 'Troponin T(ng/ML)', 'anchor_age': 'age'})
```

### Filling Some Null values with Random values

```python
import numpy as np

np.random.seed(42)

# Iterate through rows with NaN in Troponin column and replace null values with a random between  0 - 0.01
for index, row in merged_df[merged_df['Troponin T(ng/ML)'].isnull()].iterrows():
  random_value = np.random.uniform(0, 0.01)
  merged_df.loc[index, 'Troponin T(ng/ML)'] = random_value
```

```python
merged_df['Troponin T(ng/ML)'] = merged_df['Troponin T(ng/ML)'].round(4)
```

### Balancing Dataset

Trying to delete class '0' as much as possible

```python
merged_df = merged_df[~((merged_df['Troponin T(ng/ML)'] == 0) | (merged_df['Troponin T(ng/ML)'] == 0.01))]
```

```python
merged_df = merged_df[~((merged_df['abnormal'] == 0) & ((merged_df['medications'].isnull() | merged_df['diagnosis'].isnull())))]
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df.isna().sum()
```

```python
# Checking class balance
negative = merged_df['abnormal'].eq(0).sum()
positive = merged_df['abnormal'].eq(1).sum()

Difference = negative-positive # initialyy 27734 but after running next cell it is 0

print(f"Negative : {negative}")
print(f"Positive : {positive}")
print(f"Difference : {Difference}")
```

```python
class_0 = merged_df[merged_df['abnormal'] == 0]
class_0 = class_0.sort_values(by=['diagnosis', 'medications'],key=lambda col: col.map(len),ascending=False)

class_0 = class_0.iloc[Difference:]

merged_df = pd.concat([merged_df[merged_df['abnormal'] == 1], class_0], ignore_index=True)
merged_df.sort_values(by=['subject_id'])
```

```python
abnormal_column = merged_df.pop('abnormal')
merged_df['abnormal'] = abnormal_column
```
